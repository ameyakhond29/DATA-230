```python
# Question 1

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("cars.csv")

data = df[['mpg', 'cyl', 'disp', 'hp', 'wt']].dropna()

plt.figure(figsize=(10, 7))

plt.scatter(
    data['wt'], 
    data['hp'],
    s=data['disp'] * 0.5,   
    c=data['mpg'],          
)

plt.xlabel("Weight")
plt.ylabel("Horsepower")

plt.title("Cars: Weight vs Horsepower\n Color = MPG, Size = Displacement")

cbar = plt.colorbar()
cbar.set_label("MPG")

plt.grid()
plt.show()
```

```python
# Question 1 (AI Solution)

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("cars.csv")

# Create figure
fig, ax = plt.subplots(figsize=(10, 7))

# Normalize horsepower for bubble sizes
size_scale = 20
sizes = (df["hp"] / df["hp"].max()) * size_scale**2  

# Scatter plot
sc = ax.scatter(
    x=df["wt"], y=df["mpg"],
    s=sizes,
    c=df["cyl"],
    cmap="plasma",
    alpha=0.8,
    edgecolor="white",
    linewidth=0.7
)

# Titles and labels
ax.set_title("Car Fuel Efficiency vs Weight", fontsize=14, fontweight="bold", pad=15)
ax.set_xlabel("Weight (1000 lbs)", fontsize=12)
ax.set_ylabel("Miles per Gallon (mpg)", fontsize=12)
ax.grid(True, linestyle="--", alpha=0.4)

# Colorbar for cylinders
cbar = plt.colorbar(sc, ax=ax, shrink=0.9, aspect=20)
cbar.set_label("Number of Cylinders", fontsize=11)

# Custom legend for horsepower bubbles
for hp in [50, 150, 250]:
    plt.scatter([], [], 
                s=(hp / df["hp"].max()) * size_scale**2, 
                c="gray", alpha=0.6, 
                edgecolor="k", linewidth=0.7,
                label=f"{hp} HP")

legend = ax.legend(title="Horsepower", frameon=True, fontsize=10, loc="upper right")
plt.setp(legend.get_title(), fontsize=11)

plt.tight_layout()
plt.show()
```

```python
# Question 2 (AI Solution)

import numpy as np
import matplotlib.pyplot as plt

# Define x
x = np.linspace(0.5, 3.5, 100)

# Define curves
y1 = 3 + np.cos(x)
y2 = 1 + np.cos(1 + x/0.75) / 2

# Random points between y1 and y2
y3 = np.random.uniform(y1, y2, len(x))

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label="y1 = 3 + cos(x)", color="cyan")
plt.plot(x, y2, label="y2 = 1 + cos(1 + x/0.75)/2", color="green")

# Shade area between y1 and y2
plt.fill_between(x, y1, y2, color="gray", alpha=0.2, label="Region between y1 & y2")

# Random points inside shaded area
plt.scatter(x, y3, label="y3 (random between y1 & y2)", color="red", s=20, alpha=0.7)

plt.xlabel("x")
plt.ylabel("y")
plt.title("y1, y2, and random y3 values between them")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.minorticks_on()
plt.show()
```
